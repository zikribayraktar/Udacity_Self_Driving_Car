## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def draw_lines(img, lines, color=[255, 0, 0], thickness=2):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    for line in lines:
        for x1,y1,x2,y2 in line:
            cv2.line(img, (x1, y1), (x2, y2), color, thickness)

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    
    modified to return only the lines.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    #line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    #draw_lines(line_img, lines)
    return lines  #line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, α=0.8, β=1., λ=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img * α + img * β + λ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, α, img, β, λ)

#################################################################
############# ADDITIONAL FUNCTIONS THAT I DEFINED ###############
#################################################################

def find_left_right_lanes(lines):
    """This function separates left road lines from right road lines by their slope.
    'lines' input is from Hugh Transform. """
    
    negMcord = []  #list to store lines coordinates with negative slope
    posMcord = []  #list to store lines coordinates with positive slope
    mneg = []  #negative slope values
    mpos = []  #positive slope values

    # LINES AVERAGING:
    for each in lines:
        for x1,y1,x2,y2 in each:
            m =  ((y2-y1)/(x2-x1))
            #print(x1,y1,x2,y2,m)
            if m>0.0 and np.abs(m) > 0.35:
                posMcord.append([x1,y1,x2,y2])
                mpos.append(m)
            elif m<0.0 and np.abs(m) > 0.35:
                negMcord.append([x1,y1,x2,y2])
                mneg.append(m)
            else:
                dummy=1
    
    return mpos, posMcord, mneg, negMcord 
#-----------------------------------------------------------

def averaged_coord_yintercept(coord, slope):

    cord_matrix = np.asmatrix(coord)
    #print(type(cord_matrix))

    xmean = (np.mean(cord_matrix[:,0]) + np.mean(cord_matrix[:,2]))/2
    ymean = (np.mean(cord_matrix[:,1]) + np.mean(cord_matrix[:,3]))/2

    meanSlope = np.mean(slope)

    b_inters = ymean -meanSlope*xmean
    return xmean, ymean, b_inters

#------------------------------------------------------------

def max_point_compute (coords):
    AA = np.vstack(coords)  #convert to array
    y1 = AA[:,1]           # y1 coordinates
    y2 = AA[:,3]           # y2 coordinates

    x1 = AA[:,0]  
    x2 = AA[:,2]

    PoneX = np.where(x1==np.max(x1))
    PoneY = np.where(y1==np.min(y1))

    PtwoX = np.where(x2==np.max(x2))
    PtwoY = np.where(y2==np.min(y2))
    
    P1indx = PoneY[0][0]
    P2indx = PtwoY[0][0]
    
    if y1[P1indx] < y2[P2indx]:
        useX = x1[P1indx]
        useY = y1[P1indx]
    else:
        useX = x2[P2indx]
        useY = y2[P2indx]
        
    return useX, useY
#-----------------------------------------------------------
````

## Build a Lane Finding Pipeline



Build the pipeline and run your solution on all test_images. Make copies into the `test_images_output` directory, and you can use the images in your writeup report.

Try tuning the various parameters, especially the low and high Canny thresholds as well as the Hough lines parameters.

````python
# TODO: Build your pipeline that will draw lane lines on the test_images
# then save them to the test_images directory.

# USER INPUTS
dirout = 'test_images_output'
gkernel = 5   #kernel for gaussian blur
canny_low_th = 50    # canny filter low threshold
canny_high_th = 150   # canny filtr high threshold

#Hough Transform parameters:
H_rho = 2
H_theta = 1*np.pi/180 
H_th=15
H_minLineLen = 50
H_maxLineGap = 20


# PIPELINE STARTS HERE
imageName = fnames[0]   # pull the file name
imageRead = mpimg.imread("test_images/"+ imageName)   # read the image
imageIn = np.copy(imageRead)  # make a copy of the image

#Grayscale conversion:
gray_image = grayscale(imageIn)

plt.imshow(gray_image, cmap='gray')
plt.show()
plt.imsave(dirout+"/gray_image_"+imageName, gray_image, cmap='gray')

#Apply Gaussian smooting to the grayscale image:
blur_gray = gaussian_blur(gray_image, gkernel)

#Apply Canny edge detection:
edges_of_blur = canny(blur_gray, canny_low_th, canny_high_th)

plt.imshow(edges_of_blur, cmap='gray')
plt.show()
plt.imsave(dirout+"/canny_image_"+imageName, edges_of_blur)

imDim = imageIn.shape  # dimensions of the image
# vertices are hardcoded here with the assumption that the imageIn size is known ahead of time:
#vertices = np.array([[(0, imDim[0]), (450,320), (520,320), (imDim[1],imDim[0])]], dtype=np.int32)
#vertices = np.array([[(0, imDim[0]), (450,320), (520,320), (imDim[1],imDim[0])]], dtype=np.int32)
xsc = imDim[1]/960  #xscale factor for different size images
ysc = imDim[0]/540  #yscale factor for different size images

Xoffst = 40 * xsc
Yoffst = 40 * ysc

#vertices = np.array([[(0, imDim[0]), (int(xsc*450),int(ysc*320)), (int(xsc*520),int(ysc*320)), (imDim[1],imDim[0])]], dtype=np.int32)
vertices = np.array([[(Xoffst, imDim[0]-Yoffst), (int(xsc*450),int(ysc*320)), (int(xsc*520),int(ysc*320)), (imDim[1]-Xoffst,imDim[0]-Yoffst)]], dtype=np.int32)

# Select the region of interest:
masked_image = region_of_interest(edges_of_blur, vertices)

plt.imshow(masked_image, cmap='gray')
plt.show()
plt.imsave(dirout+"/masked_image_"+imageName, masked_image)

# Apply Hough Transform to find lines within the masked_image:
lines = hough_lines(masked_image, H_rho, H_theta, H_th, H_minLineLen, H_maxLineGap)

# RGB image -- blank to draw lines on
line_image = np.copy(imageIn)*0
# Draw the lines on the blank image:
draw_lines(line_image, lines, color=[255, 0, 0], thickness=2)


# Combine the RGB image with the line_image:
combo = cv2.addWeighted(imageIn, 0.8,  line_image, 0.8, 0)

plt.imshow(combo)
plt.show()
plt.imsave(dirout+"/with_lines_"+imageName, combo)

# Separate left lines (negative slope) and right lines (positive slope) by their slope:
mpos, posMcord, mneg, negMcord  = find_left_right_lanes(lines)

# Left lines: find b_intersect for line y=mx+b, where line intersects the y-axis of the image:
xmean_left, ymean_left, b_left = averaged_coord_yintercept(coord=negMcord, slope=mneg)

# Right lines: find b_intersect for line y=mx+b, where line intersects the y-axis of the image:
xmean_rite, ymean_rite, b_rite = averaged_coord_yintercept(coord=posMcord, slope=mpos)

# Coordinates that intersect the bottom of the image:
x_left_box_inters = (imDim[0] - b_left)/np.mean(mneg)
x_rite_box_inters = (imDim[0] - b_rite)/np.mean(mpos)

# Compute the maximum extention of the line:
PosUseX, PosUseY = max_point_compute(posMcord)
NegUseX, NegUseY = max_point_compute(negMcord)


line_image = np.copy(imageIn)*0

draw_lines(line_image, [[[int(x_left_box_inters),int(imDim[0]),int(NegUseX),int(NegUseY)]]], color=[0,255,0], thickness=16)
draw_lines(line_image, [[[int(x_rite_box_inters),int(imDim[0]),int(PosUseX),int(PosUseY)]]], color=[0,0,255], thickness=16)

# Combine the RGB image with the line_image:
combo = cv2.addWeighted(imageIn, 0.9,  line_image, 0.7, 0)

plt.imshow(combo)
plt.show()
plt.imsave(dirout+"/"+imageName, combo)
````

````python
def process_image(image):
    # NOTE: The output you return should be a color image (3 channel) for processing video below
    # TODO: put your pipeline here,
    # you should return the final output (image where lines are drawn on lanes)

    # TODO: Build your pipeline that will draw lane lines on the test_images
    # then save them to the test_images directory.

    # USER INPUTS
    gkernel = 5   #kernel for gaussian blur
    canny_low_th = 50    # canny filter low threshold
    canny_high_th = 150   # canny filtr high threshold

    #Hough Transform parameters:
    H_rho = 2
    H_theta = 1*np.pi/180 
    H_th=15
    H_minLineLen = 50
    H_maxLineGap = 20

    # PIPELINE STARTS HERE
    imageIn = np.copy(image)  # make a copy of the image

    #Grayscale conversion:
    gray_image = grayscale(imageIn)

    #Apply Gaussian smooting to the grayscale image:
    blur_gray = gaussian_blur(gray_image, gkernel)

    #Apply Canny edge detection:
    edges_of_blur = canny(blur_gray, canny_low_th, canny_high_th)

    imDim = imageIn.shape  # dimensions of the image
    
    #Utilize image scaling factors if the image input keeps changing shape:
    xsc = imDim[1]/960  #xscale factor for different size images
    ysc = imDim[0]/540  #yscale factor for different size images
    
    #Offset from the edges of the bottom of the image. No need to extend the mask to the far corners:
    Xoffst = 40 * xsc
    Yoffst = 40 * ysc
    vertices = np.array([[(Xoffst, imDim[0]-Yoffst), (int(xsc*450),int(ysc*320)), (int(xsc*520),int(ysc*320)), (imDim[1]-Xoffst,imDim[0]-Yoffst)]], dtype=np.int32)

    # Select the region of interest:
    masked_image = region_of_interest(edges_of_blur, vertices)
    
    # Apply Hough Transform to find lines within the masked_image:
    lines = hough_lines(masked_image, H_rho, H_theta, H_th, H_minLineLen, H_maxLineGap)

    # RGB image -- blank to draw lines on
    line_image = np.copy(imageIn)*0
    # Draw the lines on the blank image:
    draw_lines(line_image, lines, color=[255, 0, 0], thickness=2)

    # Combine the RGB image with the line_image:
    combo = cv2.addWeighted(imageIn, 0.8,  line_image, 0.8, 0)

    # Separate left lines (negative slope) and right lines (positive slope) by their slope:
    mpos, posMcord, mneg, negMcord  = find_left_right_lanes(lines)

    
    line_image = np.copy(imageIn)*0
    
    # Implement fail-safe in case no coordinates found by the Hough transform:
    if negMcord:   #if is not empty
        # Left lines: find b_intersect for line y=mx+b, where line intersects the y-axis of the image:
        xmean_left, ymean_left, b_left = averaged_coord_yintercept(coord=negMcord, slope=mneg)
 
        x_left_box_inters = (imDim[0] - b_left)/np.mean(mneg)
    
        NegUseX, NegUseY = max_point_compute(negMcord)
        
        draw_lines(line_image, [[[int(x_left_box_inters),int(imDim[0]),int(NegUseX),int(NegUseY)]]], color=[0,255,0], thickness=16)


    if posMcord:
        # Right lines: find b_intersect for line y=mx+b, where line intersects the y-axis of the image:
        xmean_rite, ymean_rite, b_rite = averaged_coord_yintercept(coord=posMcord, slope=mpos)

        x_rite_box_inters = (imDim[0] - b_rite)/np.mean(mpos)

        PosUseX, PosUseY = max_point_compute(posMcord)
        
        draw_lines(line_image, [[[int(x_rite_box_inters),int(imDim[0]),int(PosUseX),int(PosUseY)]]], color=[0,0,255], thickness=16)

    # Combine the RGB image with the line_image:
    combo = cv2.addWeighted(imageIn, 0.9,  line_image, 0.7, 0)    
    
    return combo
````
